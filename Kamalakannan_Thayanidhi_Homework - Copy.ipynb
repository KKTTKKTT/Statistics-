# Q) Chapter 3 - Ex 10

```r
library(ISLR)
cs <- Carseats
head(cs)
```

```r
#a) 
#Multiple Regression fitting 
lm.fit = lm(cs$Sales~cs$Price + cs$Urban + cs$US)
```

```r
summary(lm.fit)
```

b) The Above statistics provides informations about the relationship of the variables 

Price - Low P Value suggests Presence of relationship, Negative coefficient suggest Inverse Co-Relation 
UrbanYes - High P Vlaue suggest No relationship. 
UsYes - Low P Value suggests Presence of relationship, Positive coefficent suggests that Presence in Us Increases the sales.

c)
Model Equation - 
Y = 13.043 + -0.054 X1 + -0.021 X2 + 1.2 X3

where X1 - Price , X2 - UrbanYes, X3 - USYes

d) for Variables Price & USYes.

```r
#e) 
#fitting smaller Model 
lm.fit2 = lm(cs$Sales~cs$Price + cs$US)
summary(lm.fit2)
```

#f)
They both fit the data similarly, with linear regression from (e) fitting the data slightly better based on Adjusted r-Square.

```r
#g) 95 % Confidence Level
confint(lm.fit2)
```

```r
#h) Outliers from Model (e)
par(mfrow = c(2, 2))
plot(lm.fit2)

# The Residual vs Leverage suggests there are 1-3 outliers. 
```
